```python
unpivot5 = pd.read_csv('../data/gdp_long_2000_2022.csv')
```

```python
unpivot6 = pd.read_csv('../data/hdi_long_2000_2022.csv')
```

```python
merged_for_corr = pd.merge(unpivot5, unpivot6, on = ['Country', 'Year'])
```

```python
overall_corr = merged_for_corr[['GDP_pct', 'HDI']].corr().loc['GDP_pct', 'HDI']
overall_corr
```

```python
def corr(groupped):
    return groupped [['GDP_pct', 'HDI']].corr().iloc[0, 1]
```

```python
by_country_corr = merged_for_corr.groupby('Country').apply(corr)
by_country = by_country_corr.to_frame().reset_index()
```

```python
by_country = by_country.rename(columns = {0 : 'Correlation'})
by_country
```

```python
by_year_corr = merged_for_corr .groupby('Year').apply (corr)
by_year = by_year_corr.to_frame().reset_index()
```

```python
by_year = by_year.rename(columns = {0 : 'Correlation'})
by_year
```
